```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("general_data.csv")
```

```python
df
```

```python
df1=pd.get_dummies(data=df,drop_first=True)
```

```python
df1.isnull().sum()
```

```python
df1.fillna(method='pad',inplace=True)
```

```python
y=df1['Attrition_Yes']
```

```python
df1.drop(['StandardHours','EmployeeCount','EmployeeID','Attrition_Yes'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df1,y,test_size=0.35,random_state=25)
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
model.coef_
```

Above model is unable to predict True Negatives so using statsmodels.api we will find the important features required for the model

```python
import statsmodels.api as sm
```

```python
X1=sm.add_constant(df1)
```

```python
model=sm.Logit(y,X1)
```

```python
result=model.fit()
```

```python
result.summary()
```

Features whose p-value is greater than 0.05, needs to be removed and create a model

```python
df1
```

```python
model1.score(X1_train,y1_train)
```

```python
y1_pred=model1.predict(X1_test)
```

```python
confusion_matrix(y1_test,y1_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y1_test,y1_pred)
```
