As all records belongs has Over18 feature has yes. So dropping Over18 feature. EmployeeId and EmployeeCount are also dropped has they are just unique ids.

We are creating two models.
1. independent features contains only numerical data type
2. independent features contains both numerical and categorical

1.Considering only numerical features

```python
df_input=df[['Age', 'DistanceFromHome','Education','JobLevel','MonthlyIncome','NumCompaniesWorked','PercentSalaryHike','StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear','YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
```

```python
df_input.isnull().sum()
```

```python
df_input.dropna(inplace=True)
```

```python
Dv=np.where(df['Attrition']=='Yes',1,0)
```

```python
Dv=pd.DataFrame(Dv)
```

```python
Dv
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
model1.fit(X=df_input,y=Dv)
```

```python
model1.oob_score_
```

```python
for feature, imp in zip(df_input.columns,model1.feature_importances_):
    print(feature,imp)
```

2.Converting categorical to numerical using get_dummies

```python
modified_df=pd.get_dummies(df,drop_first=True)
```

```python
modified_df.dropna(axis=0, how='any',inplace=True)
```

```python
atrtiton=modified_df['Attrition_Yes']
```

```python
modified_df.drop(['Attrition_Yes'],axis=1,inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
model.fit(X=modified_df,y=atrtiton)
```

```python
model.oob_score_
```

```python
for feature, imp in zip(modified_df.columns,model.feature_importances_):
    print(feature,imp)
```

For both models we are recieving same accuracies
