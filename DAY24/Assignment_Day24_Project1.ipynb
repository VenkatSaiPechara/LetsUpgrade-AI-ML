```python
import pandas as pd
import numpy as np
```

```python
titanic_train=pd.read_csv('train.csv')
```

```python
titanic_train
```

```python
titanic_train.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
encoded_sex_train=label_encoder.fit_transform(titanic_train["Sex"])
```

```python
train1=pd.DataFrame([encoded_sex_train,titanic_train["Age"],titanic_train["Fare"]]).T
```

```python
train1
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X=train1,y=titanic_train['Survived'])
```

```python
print("Accuracy for model using features like sex, age and fare is ",model.score(X=train1,y=titanic_train['Survived']))
```

```python
titanic_test=pd.read_csv('test.csv')
```

```python
titanic_test
```

```python
titanic_test.isnull().sum()
```

```python
encoded_sex_test=label_encoder.fit_transform(titanic_test["Sex"])
```

```python
test1=pd.DataFrame([encoded_sex_test,titanic_test["Age"],titanic_test["Fare"]]).T
```

```python
test1
```

```python
test1['Survived']=model.predict(test1)
```
