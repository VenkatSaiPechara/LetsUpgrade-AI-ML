```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("train.csv")
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
embarked=lb.fit_transform(df['Embarked'])
sex=lb.fit_transform(df['Sex'])
```

```python
df['Embarked']=embarked
df['Sex']=sex
```

```python
df
```

```python
Dv=df['Survived']
```

```python
df.drop(['Survived'],axis=1,inplace=True)
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
model=BernoulliNB()
```

```python
model.fit(df,Dv)
```

```python
model.score(df,Dv)
```
