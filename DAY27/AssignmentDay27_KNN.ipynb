```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('train.csv')
```

```python
df
```

```python
df.drop(['Cabin','PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
df
```

```python
a=pd.get_dummies(df[['Embarked','Sex']])
```

```python
df=pd.concat([df,a],axis=1)
```

```python
df
```

```python
df.drop(['Sex','Embarked'],axis=1,inplace=True)
```

```python
df
```

```python
Dv=df['Survived']
df.drop(['Survived'],axis=1,inplace=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(df,Dv,test_size=0.25)
```

```python
model=KNeighborsClassifier()
```

```python
accuracy=[0]
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
for i in range(1,len(Xtrain)):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(Xtrain,ytrain)
    ypred=model.predict(Xtest)
    t=accuracy_score(ytest,ypred)
    print('Accuracy score of the model is ',t)
    accuracy.append(t)
    print("Confusion matrix when n_neidbhors is:",i)
    print(confusion_matrix(ytest,ypred))
    
```

```python
max=accuracy[0]
index=0
```

```python
for i in range(len(accuracy)):
    if max<accuracy[i]:
        max=accuracy[i]
        index=i
```

```python
print("Best K-nearest neighbor value is ", index)
```
